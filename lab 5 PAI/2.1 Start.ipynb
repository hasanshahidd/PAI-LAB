```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def show_image(img, title="Image"):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB) if len(img.shape) == 3 else img, cmap="gray")
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
img = cv2.imread( r"C:\Users\Admin\Pictures\Screenshots\2.1.png") 
show_image(img, "Original Image")
```

```python

cv2.imshow("Displayed Image", img)
cv2.waitKey(200)
cv2.destroyAllWindows()
```

```python

cv2.imwrite("saved_image.jpg", img)
print("Image Saved")
```

```python

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_image(gray_img, "Grayscale Image")
```

```python

brighter_img = cv2.add(img, np.ones(img.shape, dtype=np.uint8) * 50)
show_image(brighter_img, "Brighter Image")
```

```python
mask = np.zeros(img.shape[:2], dtype=np.uint8)
cv2.rectangle(mask, (50, 50), (200, 200), 255, -1)
show_image(mask, "Mask")
```

```python
bitwise_and = cv2.bitwise_and(img, img, mask=mask)
show_image(bitwise_and, "Bitwise AND with Mask")
bitwise_or = cv2.bitwise_or(img, img, mask=mask)
show_image(bitwise_or, "Bitwise OR with Mask")
bitwise_xor = cv2.bitwise_xor(img, img, mask=mask)
show_image(bitwise_xor, "Bitwise XOR with Mask")
bitwise_not = cv2.bitwise_not(img, mask=mask)
show_image(bitwise_not, "Bitwise NOT with Mask")
```
